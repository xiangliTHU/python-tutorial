# 图，子图，轴，记号

之前我们使用到了图和轴，但是没有具体说明图和轴的用法。在 `matplotlib` 中，图可以包换多个子图，每个子图中包含多个轴。

当我们使用 `plot` 函数的时候，`matplotlib` 调用 `gca()` 得到现有的轴，然后调用 `gcf()` 得到现有的图像；如果没有现有的图像，那么 `matplotlib` 使用 `figure()` 产生一个新的图像，或者严格说来，产生一个 `subplot(111)`。

## 图

图像以 `Figure #` 为窗口标题，并且数字从 1 开始，`figure()` 函数的主要参数如下：

参数 | 默认值 | 描述
---|---|---
`num`|`1`| 图号
`figsize`|`figure.figsize`| 图大小（宽，高）（单位英寸）
`dpi`|`figure.dpi`| 分辨率（每英寸所打印的点数）
`facecolor`|`figure.facecolor`| 背景颜色
`edgecolor`|`figure.edgecolor`| 边界颜色
`frameon` |`True`| 是否显示图框架

```python
import matplotlib as mpl
import matplotlib.pylab as plt
import numpy as np
```

创建一副新的图像：

```python
%matplotlib inline

# 产生一副新的图像，标号默认从1开始
f = plt.figure()

a = np.linspace(-5., 5.)

b = a ** 2

p = plt.plot(a, b)
```

修改默认参数值：

```python
# 产生一副新的图像
f = plt.figure(figsize=(9,6), dpi=100)

a = np.linspace(-5., 5.)
b = a ** 2
p = plt.plot(a, b)
```

如果不指定标号，那么图的标号将从1开始逐渐增加，也可以指定图像的标号：

```python
# 产生 figure 6
f = plt.figure(6)

a = np.linspace(-5., 5.)
b = a ** 2
p = plt.plot(a, b)
```

## 子图

在 **Matlab** 中可以使用 `subplot` 产生子图，`matplotlib` 也不例外，我们可以使用 `plt.subplot` 来产生子图：

这里我们使用 `subplot2grid` 函数来产生空的：

```python
f = plt.figure(figsize=(12,9), dpi=80)

def make_ticklabels_invisible(fig):
    for i, ax in enumerate(fig.axes):
        ax.text(0.5, 0.5, "ax%d" % (i+1), va="center", ha="center")
        for tl in ax.get_xticklabels() + ax.get_yticklabels():
            tl.set_visible(False)

ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax4 = plt.subplot2grid((3,3), (2,0))
ax5 = plt.subplot2grid((3,3), (2,1))

plt.suptitle("subplot2grid")
make_ticklabels_invisible(plt.gcf())
```
