##简单的数学运算

整数相加，得到整数：

```python
2 + 2
```

浮点数相加，得到浮点数：

```python
2.0 + 2.5
```

整数和浮点数相加，得到浮点数：

```python
2 + 2.5
```

##字符串 String

字符串的生成，单引号与双引号是等价的：

```python
s = "hello world"
s
```

```python
s = 'hello world'
s
```

三引号用来输入包含多行文字的字符串：

```python
s = """hello
world"""
print s
```

```python
s = '''hello
world'''
print s
```

字符串的加法：

```python
s = "hello" + " world"
s
```

字符串索引：


```python
s[0]
```

```python
s[-1]
```

```python
s[0:5]
```

字符串的分割：

```python
s = "hello world"
s.split()
```

查看字符串的长度：

```python
len(s)
```
