# Python 数据类型

##常用数据类型 Common Data Types

| 类型| 例子|
| -----  | ----- |
| 整数 | `-100` |
| 浮点数 | `3.1416` |
| 字符串 | `'hello'` |
| 列表 | `[1, 1.2, 'hello']` |
| 字典 | `{'dogs': 5, 'pigs': 3}`|
| Numpy数组 | `array([1, 2, 3])`
